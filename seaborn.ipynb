```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7]
temp=[30,35,40,45,32,37,47]
sns.lineplot(x='days',y='temp')
```

```python
days=[1,2,3,4,5,6,7]
temp=[30,35,40,45,32,37,47]
temp_df=pd.DataFrame({'days':days,'temp':temp})
sns.lineplot(x='days',y='temp',data=temp_df)
plt.show()
```

```python
#Kaggle.com
#github
#seaborn.load.dataset
tips_df=sns.load_dataset('tips')
tips_df
```

```python
tips_df.shape
```

```python
sns.lineplot(x='size',y='total_bill',data=tips_df,hue='sex',style='sex')
plt.plot()
```

```python
#palette
plt.figure(figsize=(10,6))
sns.set(style='darkgrid')
sns.lineplot(x='size',y='total_bill',data=tips_df,hue='sex',style='sex',palette='inferno',dashes=False,markers=['o','<'],legend='brief')
plt.title('Line plot for sizes vs Total Bill',fontsize=20)
plt.xlabel('size',fontsize=20)
plt.ylabel('Total Bill',fontsize=20)
plt.plot()
```
